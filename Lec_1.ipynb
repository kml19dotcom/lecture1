```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Wczytanie danych
df = pd.read_csv('Salary Data (2).csv')

# Usunięcie ewentualnych braków danych dla pewności
df = df.dropna(subset=['Years of Experience', 'Salary', 'Education Level'])

# Stylizacja wykresu
sns.set_theme(style="whitegrid")
plt.figure(figsize=(12, 7))

# Stworzenie wykresu punktowego
sns.scatterplot(
    data=df,
    x='Years of Experience',
    y='Salary',
    hue='Education Level',
    palette='viridis',
    s=80,
    alpha=0.7
)

plt.title('Zależność wynagrodzenia od lat doświadczenia i wykształcenia', fontsize=16, pad=15)
plt.xlabel('Lata doświadczenia', fontsize=12)
plt.ylabel('Wynagrodzenie (USD)', fontsize=12)
plt.legend(title='Poziom wykształcenia')

# Zapisanie wykresu do pliku
plt.tight_layout()
plt.savefig('salary_vs_experience.png', dpi=300)
```
